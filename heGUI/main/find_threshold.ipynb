```python
from skimage import img_as_ubyte
from skimage.measure import find_contours
from skimage.transform import resize
import napari.layers
from skimage.transform import rotate
from matplotlib import pyplot as plt
from skimage.morphology import skeletonize
from skimage.measure import regionprops
from skimage.measure import label
import numpy as np
from skimage import io
```

```python
he = io.imread("/Users/yokote.k/Desktop/MIBI/HE_GUI/heGUI/data/MIBI_CM21-07_he.jpg")
```

```python
plt.figure(figsize=(20,20))
plt.imshow(he)
```

```python
new_image = (he[..., 0] > 160) * (he[..., 1] > 100) * (he[..., 2] < 180)

label_im = label(new_image.astype(np.uint8))
regions = regionprops(label_im)
regions.sort(key=lambda x: x.area, reverse=True)
regions = regions[:4]

fig, ax = plt.subplots(figsize=(20,20))
ax.imshow(new_image, cmap=plt.cm.gray)

for contour in regions:
    ax.plot(contour.coords[:, 1], contour.coords[:, 0], linewidth=2)

print(regions[0].coords)
```
